````python
# Creating function 
def countOne(x):
    try: 
        ones = x[x==1]
        count = ones.size
        return count
    except not isinstance(x, np.ndarray):
        return print('invalid input')
    


countOne(er_stat_int_array)
````

# Loading data frames into SQL data base
- using sqlite

````python
import sqlite3 
conn = sqlite3.connect('data/sql/tcga_brca_sql')
conn.commit()
````

````python
clinMatchDF.info()
````

````python
# Removing duplicated colnames
clinMatchDF.drop('Integer_Stage', axis='columns', inplace=True)
clinMatchDF.info()
````

````python
# Adding tables
clinMatchDF.to_sql("clinical", con=conn, if_exists="replace")
cpmDF.to_sql("cpm", con=conn, if_exists="replace")
countsDF.to_sql("counts", con=conn, if_exists="replace")
````

## Doing Basic Queries

````python
query = 'SELECT INTEGER_STAGE, OS_EVENT, OS_TIME from clinical where OS_EVENT = 1'
queryDF = pd.read_sql_query(query, con=conn)
queryDF.head()
````
